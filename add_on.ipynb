```python
import os
from mistralai import Mistral

# Replace 'your_api_key' with your actual API key
api_key = 'MISTRAL_API_KEY'
client = Mistral(api_key=api_key)
```

```python
def analyze_sentiment(review):
    prompt = f"Analyze the sentiment of the following drug review and classify it as Positive, Negative, or Neutral:\n\n{review}"
    response = client.chat.complete(
        model='mistral-large-latest',  # Specify the model
        messages=[{"role": "user", "content": prompt}],
        temperature=0.3  # Adjust temperature for response randomness
    )
    return response.choices[0].message.content.strip()
```

```python
# Example list of reviews
reviews = [
    "This medication worked wonders for my condition.",
    "I experienced severe side effects with this drug.",
    "The treatment was okay, not much improvement noticed."
]

# Analyze sentiments
for review in reviews:
    sentiment = analyze_sentiment(review)
    print(f"Review: {review}\nSentiment: {sentiment}\n")
```
